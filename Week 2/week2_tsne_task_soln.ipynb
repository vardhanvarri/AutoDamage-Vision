## Objective
Visualize high-dimensional image data in 2D using **t-SNE**.

**Note:** t-SNE is for visualization only.


```python
# TODO: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
```

## Step 1: Load Dataset

```python
# TODO: Load digits dataset again
digits = load_digits()
x = digits.images
y = digits.target
x.shape,y.shape,y
```

## Step 2: Preprocess

```python
# TODO: Flatten images and normalize pixel values

X = x.reshape(len(x),-1)
scalar = StandardScaler()
x_scal = scalar.fit_transform(X)
```

## Step 3: Apply t-SNE

```python
# TODO: Initialize TSNE with n_components=2
# Fit and transform data
tsne = TSNE(n_components=2,init='random')
x_tsne = tsne.fit_transform(x_scal)
x_tsne
```

## Step 4: Visualization

```python
# TODO: Scatter plot embeddings
# Color points using labels
plt.figure(figsize=(8,6))
scatter = plt.scatter(x_tsne[:,0],x_tsne[:,1],c=y,cmap='tab10',s=10)
plt.colorbar(scatter, ticks=range(10))
plt.title("t-SNE Visualization (Digits Dataset)")
plt.xlabel("TSNE Component 1")
plt.ylabel("TSNE Component 2")
plt.show()
```
