## Step 3: Apply PCA

```python
# TODO: Initialize PCA with n_components = 50
# Fit PCA on flattened images
pca = PCA(n_components=20)
X_pca = pca.fit_transform(x)
```

## Step 4: Reconstruction

```python
# TODO: Transform and inverse_transform using PCA
X_reconstructed = pca.inverse_transform(X_pca)  # inverse_transform
```

## Step 5: Visualization

```python
# TODO: Plot original vs reconstructed images using matplotlib

plt.subplot(1,2,1)
plt.imshow(x[0].reshape(8,8), cmap="gray")
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(X_reconstructed[0].reshape(8,8), cmap="gray")
plt.title("Reconstructed")

plt.show()
```
